# Gradient descent for a function with one variable

## Import libraries

```python
import numpy as np
from matplotlib import pyplot as plt
```

## Practice

Finding the local minimum of the function $f(x)=x^4-10x+3$ using the gradient descent method.

Defining the function first.

```python
def f(x):
   return x**4 - 10*x + 3
```

Creating the graph of the function.

```python
x_array = [x for x in np.arange(-5, 5, 0.1)]
f_array = [f(x) for x in np.arange(-5, 5, 0.1)]

plt.figure(figsize=(8,8))

plt.axvline(x=0, c = 'pink')
plt.axhline(y=0, c = 'violet')
plt.plot(x_array, f_array);
```

Calculating the derivative of the function $f(x)$:
$f'(x)=4x^3-10$.



```python
def grad_f (x):
    return 4*(x**3) - 10
```

The implement of the method in a form of a function.

```python
def gradient_descent(x_start, learning_rate, epsilon, num_iterations):
    x_curr = x_start

    for i in range(num_iterations):
      x_new = x_curr - learning_rate * grad_f(x_curr)

      if abs(x_new - x_curr) < epsilon:
        return x_curr

      x_curr = x_new

    return x_curr
```

Starting from the point $x_0 = 4$ and go with a small step $\eta = 0.01$.

```python
x0 = 5
eta = 0.01
epsilon = 0.001
iterations = 10000

x_res = gradient_descent(x0, eta, epsilon, iterations)
x_res
```

Marking the point in the graph.

```python
plt.figure(figsize=(8,8))

plt.axvline(x=0, c = 'pink')
plt.axhline(y=0, c = 'pink')
plt.plot(x_array, f_array)

plt.scatter(x_res, f(x_res), marker='o', c='red');
```
